```python
import pandas as pd
titanic_survival=pd.read_csv("titanic_train.csv")
# series_fare=titanic_survival['Fare']
id_fare=titanic_survival[["PassengerId","Fare"]]
print(id_fare.head())
```

```python
import matplotlib.pyplot as plt
#什么也不传，一张空图
plt.plot()
plt.show()
```

```python
first_twelve=id_fare[0:12]
#plot(x轴的数据，y轴的数据)
plt.plot(first_twelve['PassengerId'],first_twelve['Fare'])
plt.show()
```

```python
#plot(x轴的数据，y轴的数据)
plt.plot(first_twelve['PassengerId'],first_twelve['Fare'])
#x轴的数字逆时针旋转45度
plt.xticks(rotation=45)
plt.show()
```

```python
#plot(x轴的数据，y轴的数据)
plt.plot(first_twelve['PassengerId'],first_twelve['Fare'])
#x轴的数字逆时针旋转45度
plt.xticks(rotation=45)
#x轴标签
plt.xlabel("PassengerId")
#y轴标签
plt.ylabel("Age")
plt.title("This is so easy")
plt.show()
```
