```python
#画多个子图，空图
import matplotlib.pyplot as plt
fig = plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,4)
plt.show()
```

```python
import numpy as np
#画多个子图
fig=plt.figure(figsize=(3,6))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)

ax1.plot(np.random.randint(1,5,5),np.arange(5))
ax2.plot(np.arange(10)*3,np.arange(10))
plt.show()
```

```python
#同一个图画多条折线
fig=plt.figure(figsize=(6,3))

plt.plot(np.random.randint(1,5,5),np.arange(5),c="r")
plt.plot(np.arange(10)*3,np.arange(10),c="blue")

plt.show()
```

```python
import pandas as pd
titanic_survival=pd.read_csv("titanic_train.csv")
# series_fare=titanic_survival['Fare']
id_fare=titanic_survival[["PassengerId","Fare"]]
# print(id_fare.head())
#同一个图画多条折线
fig=plt.figure(figsize=(10,3))
colors=['red','blue','green','orange','black']
for i in range(5):
    start_index=i*12
    end_index=(i+1)*12
    subset=id_fare[start_index:end_index]
    label=str(1948+i)
#     print(colors[i])
    plt.plot(subset["PassengerId"],subset["Fare"],c=colors[i],label=label)
plt.legend(loc="best")
plt.show()
```
